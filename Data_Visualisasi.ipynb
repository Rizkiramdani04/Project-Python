##Types Data Visualisasi

```python
import pandas as pd
data=pd.read_csv('/content/sample_data/california_housing_test.csv')
data.head()
```

### Line Plots

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.plot(data['total_bedrooms'],marker='o',linestyle='-')
plt.title('Line Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.show()
```

```python
x=data['total_bedrooms']
y=data['population']
plt.bar(x,y, color='yellow',edgecolor='blue',
        linewidth=2)
plt.title('Bar Plot')
plt.xlabel('Total Bedrooms')
plt.ylabel('Population')
plt.show()
```

Kustomisasi yang tersedia untuk Bar Chart :
- color: Untuk wajah bar
- edgecolor: Warna tepi bilah
- linewidth: Lebar tepi bilah
- width: Lebar bilah

```python
x=data['households']
plt.hist(x)
plt.title('Histogram')
plt.show()
```

Kustomisasi yang tersedia untuk Histogram

- bins: Jumlah tempat sampah dengan lebar yang sama
- color: Untuk mengubah warna wajah
- edgecolor: Warna tepi
- linestyle: Untuk garis tepi
-  alpha: nilai campuran, antara 0 (transparan) dan 1 (buram)

```python
x=data['total_rooms']
plt.hist(x,bins=20,
         color='blue',
         edgecolor='yellow',
         linestyle='--',
         alpha=0.3)
plt.title('Histogram')
plt.show()
```

### ScatterPlot

```python
x=data['households']
y=data['total_rooms']
plt.scatter(x,y)
plt.title('Scatter Plot')
plt.xlabel('Households')
plt.ylabel('Total_Rooms')
plt.show()
```

Grafik ini bisa lebih bermakna jika kita bisa menambahkan warna dan juga mengubah ukuran titik-titiknya. Kita dapat melakukan ini dengan menggunakan parameter c dan s masing-masing dari fungsi scatter.

Kustomisasi yang tersedia untuk scatter adalah 

- s: ukuran penanda (bisa skalar atau array dengan ukuran sama dengan ukuran x atau y)
- c: warna urutan warna untuk spidol
- marker: gaya penanda
- linewidths: lebar batas penanda
- edgecolor: warna batas penanda
- alpha: nilai campuran, antara 0 (transparan) dan 1 (buram)


Kita juga dapat menampilkan bilah warna menggunakan metode colorbar()

```python
x=data['households']
y=data['total_rooms']
plt.scatter(x,y,c=data['population'], s=data['total_rooms'],marker='D',
            alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('Households')
plt.ylabel('Total_Rooms')
plt.colorbar()
plt.show()
```
