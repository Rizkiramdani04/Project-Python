```python
import pandas as pd
```

```python
dataset=pd.read_csv('/content/sample_data/california_housing_train.csv')
dataset.head()
```

```python
#total_rooms dikurangi total bedrooms
other_rooms=dataset['total_rooms']-dataset['total_bedrooms']
other_rooms
```

```python
#membagi nilai populasi dengan household
kepadatan=dataset['population'].div(dataset['households'])
kepadatan
```

```python
#menambahkan kolom
dataset['other_rooms']=other_rooms
dataset
```

```python
#menambahkan kolom menggunakan assign
dataset=dataset.assign(kepadatan=kepadatan)
dataset
```

```python
#menambahkan menggunakan insert sesuai dengan posisi yang kita inginkan
dataset.insert(3,'price_income_ratio', dataset['median_house_value'].div(dataset['median_income']))
dataset
```

```python
#kolom kepadatan ruang
kepadatan_ruang=dataset['total_rooms']/ dataset['households']
#menambahkan variabel kepadatan_ruang dengan dipangkatkan
dataset.assign(kepadatan_ruang_kuadrat=kepadatan_ruang.pow(2))
dataset
```

```python
dataset=sns.load_dataset('mpg')
dataset
```

```python
dataset['origin'].unique()
```

```python
#merubah nilai usa di kolom origin dengan negara amerika
dataset.loc[dataset['origin']=='usa','origin']='amerika'
dataset
```

```python
#membuat kolom baru
dataset.loc[dataset['origin']=='amerika','origin_baru']='USA'
dataset
```

```python
#merubah value japan menjadi asia
dataset['origin_baru']=dataset['origin'].replace('japan','asia')
dataset.query("origin_baru=='asia'")
```

```python
#merubah nilai menggunakan map
dataset['origin_baru']=dataset['origin'].map({'amerika':'Amerika',
                                              'asia':'Asia',
                                              'europe':'Eropa'})
dataset
```

```python
#merubah nama kolom
dataset=dataset.rename(columns={'origin':'Negara'})
dataset
```

```python
#menghapus kolom name
dataset=dataset.drop(columns='name')
dataset
```

```python
#menghapus kolom origin_baru
dataset.drop(columns='origin_baru',inplace=True)
dataset
```
