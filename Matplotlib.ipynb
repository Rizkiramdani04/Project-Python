##Pyplot

```python
import matplotlib.pyplot as plt
#nilai
x=[1,2,3,4,5]
y=[2,3,4,1,4]
#plotting data
plt.plot(x,y)
plt.show()
```

##Menambahkan Title

```python
plt.plot(x,y)
plt.title("Sample Plot",fontsize=21,color='green')
plt.show()
```

##Menambahkan label sumbu x dan sumbu y
Dalam istilah awam, label X dan label Y adalah sebutan yang diberikan masing-masing untuk sumbu X dan sumbu Y. Ini dapat ditambahkan ke grafik dengan menggunakan metode xlabel() dan ylabel().
syntax:

matplotlib.pyplot.xlabel(xlabel, fontdict=None, labelpad=None, **kwargs)

matplotlib.pyplot.ylabel(ylabel, fontdict=None, labelpad=None, **kwargs)

```python
plt.plot(x,y)
plt.title("Sample Plot",fontsize=18,color='blue')
#label sumbu x
plt.xlabel("Sumbu X")
#label Sumbu y
plt.ylabel('Sumbu Y')
plt.show()
```

## Menetapkan Batas dan label Centang
Kita mungkin telah melihat bahwa Matplotlib secara otomatis menetapkan nilai dan penanda (titik) dari sumbu X dan Y, namun, dimungkinkan untuk mengatur batas dan penanda secara manual. Fungsi xlim() dan ylim() digunakan untuk mengatur batas sumbu X dan sumbu Y masing-masing. Demikian pula, fungsi xticks() dan yticks() digunakan untuk menyetel label centang.

```python
plt.plot(x,y)
plt.title('Sample Data',fontsize=12,color='black')
plt.ylabel('Sumbu Y')
plt.xlabel('Sumbu X')
#menseting batasan di sumbu y
plt.ylim(0,8)
plt.xticks(x,labels=['A','B','C','D','E'])
plt.show()
```

## Menambahkan legends
Legend adalah area yang menggambarkan elemen grafik. Secara sederhana, ini mencerminkan data yang ditampilkan dalam sumbu Y. Biasanya muncul sebagai kotak yang berisi sampel kecil dari setiap warna pada grafik dan deskripsi kecil tentang arti data ini.

Atribut bbox_to_anchor=(x, y) dari fungsi legend() digunakan untuk menentukan koordinat legend, dan atribut ncol mewakili jumlah kolom yang dimiliki legend. Nilai defaultnya adalah 1.

syntax:

matplotlib.pyplot.legend([“name1”, “name2”], bbox_to_anchor=(x, y), ncol=1)

```python
import matplotlib.pyplot as plt


# initializing the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# plotting the data
plt.plot(x, y)

# Adding title to the plot
plt.title("Grafik Linear", fontsize=25, color="blue")

# Adding label on the y-axis
plt.ylabel('Sumbu Y')

# Adding label on the x-axis
plt.xlabel('Sumbu X')

# Setting the limit of y-axis
plt.ylim(0, 80)

# setting the labels of x-axis
plt.xticks(x, labels=["one", "two", "three", "four"])

# Adding legends
plt.legend(["GFG"])

plt.show()
```

## Figure Class
class matplotlib.figure.Figure(figsize=None, dpi=None, facecolor=None, edgecolor=None, linewidth=0.0, frameon=None, subplotpars=None, tight_layout=None, constrained_layout=None)

```python
# Python program to show pyplot module
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

# initializing the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# Creating a new figure with width = 7 inches
# and height = 5 inches with face color as
# green, edgecolor as red and the line width
# of the edge as 7
fig = plt.figure(figsize =(7, 5), facecolor='g',
				edgecolor='b', linewidth=7)

# Creating a new axes for the figure
ax = fig.add_axes([1, 1, 1, 1])

# Adding the data to be plotted
ax.plot(x, y)

# Adding title to the plot
plt.title("Grafik Linear", fontsize=25, color="yellow")

# Adding label on the y-axis
plt.ylabel('Sumbu Y')

# Adding label on the x-axis
plt.xlabel('Sumbu X')

# Setting the limit of y-axis
plt.ylim(0, 80)

# setting the labels of x-axis
plt.xticks(x, labels=["one", "two", "three", "four"])

# Adding legends
plt.legend(["GFG"])

plt.show()
```

##Axes Class
class Axes adalah unit paling dasar dan fleksibel untuk membuat sub-plot. Gambar tertentu mungkin berisi banyak sumbu, tetapi sumbu tertentu hanya dapat ada dalam satu gambar. Fungsi axes() membuat objek axes.

syntax:

axes([left, bottom, width, height])

```python
# Python program to show pyplot module
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

# initializing the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

fig = plt.figure(figsize = (5, 4))

# Adding the axes to the figure
ax = fig.add_axes([1, 1, 1, 1])

# plotting 1st dataset to the figure
ax1 = ax.plot(x, y)

# plotting 2nd dataset to the figure
ax2 = ax.plot(y, x)

# Setting Title
ax.set_title("Grafik Linear")

# Setting Label
ax.set_xlabel("Sumbu X")
ax.set_ylabel("Sumbu Y")

# Adding Legend
ax.legend(labels = ('garis 1', 'garis 2'))

plt.show()
```

## Multiple Plot (Plot lebih dari 1)

Kita telah mempelajari tentang komponen dasar graf yang dapat ditambahkan sehingga dapat menyampaikan informasi lebih lanjut. Salah satu caranya adalah dengan memanggil fungsi plot berulang kali dengan kumpulan nilai yang berbeda seperti yang ditunjukkan pada contoh di atas. Sekarang mari kita lihat cara memplot banyak grafik menggunakan beberapa fungsi dan juga cara memplot subplot.

using add_axes()

```python
# Python program to show pyplot module
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

# initializing the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# Creating a new figure with width = 5 inches
# and height = 4 inches
fig = plt.figure(figsize =(5, 4))

# Creating first axes for the figure
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Creating second axes for the figure
ax2 = fig.add_axes([1, 0.1, 0.8, 0.8])

# Adding the data to be plotted
ax1.plot(x, y)
ax2.plot(y, x)

plt.show()
```

syntax:

subplot(nrows, ncols, index, **kwargs)

subplot(pos, **kwargs)

subplot(ax)

```python
import matplotlib.pyplot as plt


# initializing the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]


# Creating figure object
plt.figure()

# adding first subplot
plt.subplot(121)
plt.plot(x, y)

# adding second subplot
plt.subplot(122)
plt.plot(y, x)
```

```python
import matplotlib.pyplot as plt


# initializing the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# Creating the figure and subplots
# according the argument passed
fig, axes = plt.subplots(1, 2)

# plotting the data in the
# 1st subplot
axes[0].plot(x, y)

# plotting the data in the 1st
# subplot only
axes[0].plot(y, x)

# plotting the data in the 2nd
# subplot only
axes[1].plot(x, y)
```

Fungsi ini membuat objek sumbu di lokasi tertentu di dalam kisi dan juga membantu merentangkan objek sumbu di beberapa baris atau kolom. Dengan kata yang lebih sederhana, fungsi ini digunakan untuk membuat beberapa grafik dalam gambar yang sama.

syntax : Plt.subplot2grid(shape, location, rowspan, colspan)

```python
import matplotlib.pyplot as plt

# initializing the data
x = [10, 20, 30, 40]
y = [20, 25, 35, 55]

# adding the subplots
axes1 = plt.subplot2grid (
(7, 1), (0, 0), rowspan = 2, colspan = 1)

axes2 = plt.subplot2grid (
(7, 1), (2, 0), rowspan = 2, colspan = 1)

# plotting the data
axes1.plot(x, y)
axes2.plot(y, x)
```

- color: warna garis
- linewidth: modifikasi width dari garis
- marker: style untuk titik koordinat
- markersize: mengatur ukuran marker
- linestyle: style untuk line yang di plot
